$y = \frac{a}{x+b}+c$  THUS $ln(y) = ln(\frac{a}{x+b}+c)$     HENCE     $ln(y) = A - Bln(x+b)$

```python
#LOG TRANSFORM 

#define inverse polynomial 
def inverse_polynomial(x, a, b, c):
    return a / (x + b) + c

#extarct data
counties = [state_data["nodes"] for state_data in graph_data.values()]
avg_success_rates = [state_data["avg_success_rate"] for state_data in graph_data.values()]
states = list(graph_data.keys())

#apply log transformations 
log_counties = np.log(counties)
log_avg_success_rates = np.log(avg_success_rates)

#fit model to original space 
params_poly, _ = curve_fit(inverse_polynomial, counties, avg_success_rates, maxfev=10000)
a_poly, b_poly, c_poly = params_poly

#generate in normal space then log transform them 
pred_poly = inverse_polynomial(counties, a_poly, b_poly, c_poly)
log_pred_poly = np.log(pred_poly)

#transform log-log space for plotting 
x_vals = np.linspace(min(counties), max(counties), 500)
y_vals_poly = inverse_polynomial(x_vals, a_poly, b_poly, c_poly)

#generate x values for plotting the fitted curve 
log_x_vals = np.log(x_vals)
log_y_vals_poly = np.log(y_vals_poly)

#plot log-log transformed data and fits
plt.figure(figsize=(10, 6))
plt.scatter(log_counties, log_avg_success_rates, color='blue', label="Log-Log Data Points")
plt.plot(log_x_vals, log_y_vals_poly, "g--", label="Log-Log Inverse Polynomial Fit")

# Formula text for inverse polynomial (in original space)
formula_text_poly = f"ln(y) = {np.log(a_poly):.2f} - ln(x + {b_poly:.2f})"
#formula_text_poly = f"y = {a_poly:.2f} / (x + {b_poly:.2f}) + {c_poly:.2f}"
plt.text(0.05, 0.95, formula_text_poly, transform=plt.gca().transAxes, fontsize=12, verticalalignment='top', color="green")

plt.ylim(-10, -5)
plt.xlabel("Log(Number of Counties)")
plt.ylabel("Log(Average Success Rate)")
plt.title("Log-Log Inverse Polynomial Fit")
plt.legend()
plt.show()
```

MEAN SQUARED ERROR (MSE) $MSE = \frac{1}{n} \sum_{i=1}^{n}(ln(y_i)-ln(\hat{y_i}))^2$

```python
from sklearn.metrics import mean_squared_error

mse_poly_log = mean_squared_error(log_avg_success_rates, log_pred_poly)
print(f"Log-Log Inverse Polynomial Model: MSE = {mse_poly_log:.4f}")
```

$R^2= 1- \frac{\sum(y-\hat{y})^2}{\sum(y-\bar{y})^2}$ 

```python
from sklearn.metrics import r2_score

r2_poly_log = r2_score(log_avg_success_rates, log_pred_poly)
print(f"Log-Log Inverse Polynomial Model: R = {r2_poly_log:.4f}")
```

MEAN ABSOLUTE ERROR $MAE = \frac{1}{n}\sum_{i=1}^{n}|ln(y_i)-ln(\hat{y_i})|$

```python
from sklearn.metrics import mean_absolute_error

mae_poly_log = mean_absolute_error(log_avg_success_rates, log_pred_poly)
print(f"Log-Log Inverse Polynomial Model: MAE = {mae_poly_log:.4f}")
```
