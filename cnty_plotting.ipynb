```python
#import packages 
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#counties vs success rate
counties = [state_data["counties"] for state_data in cnty_data.values()]
avg_success_rates = [state_data["avg_success_rate"] for state_data in cnty_data.values()]
states = list(cnty_data.keys())

plt.figure(figsize=(10, 6))
plt.scatter(counties, avg_success_rates)

plt.xlabel("Number of Counties")
plt.ylabel("Average Success Rate (%)")
plt.title("Average Success Rate vs. Number of Counties")

for i, state in enumerate(states):
    plt.annotate(state, (counties[i], avg_success_rates[i]), fontsize=8, alpha=0.7)

plt.show()
```

```python
#inverse exponential fit 
def inverse_exponential(x, a, b, c):
    return a * np.exp(-b * x) + c

params, _ = curve_fit(inverse_exponential, counties, avg_success_rates, maxfev=10000)
a, b, c = params  # Unpack parameters

plt.figure(figsize=(10, 6))
plt.scatter(counties, avg_success_rates, color='blue', label="States")

x_vals = np.linspace(min(counties), max(counties), 500)
plt.plot(x_vals, inverse_exponential(x_vals, a, b, c), "r--", label="Inverse Exponential")

formula_text = f"y = {a:.2f} * exp(-{b:.2f} * x) + {c:.2f}"
plt.text(0.05, 0.95, formula_text, transform=plt.gca().transAxes, fontsize=12, verticalalignment='top', color="red")

plt.xlabel("Number of Counties")
plt.ylabel("Average Success Rate (%)")
plt.title("Inverse Exponential Fit of Average Success Rate vs. Number of Counties")
plt.legend()
plt.show()
```

```python
#inverse polynomial fit 
def inverse_polynomial(x, a, b, c):
    return a / (x + b) + c

params, _ = curve_fit(inverse_polynomial, counties, avg_success_rates, maxfev=10000)
a, b, c = params  # Unpack parameters

plt.figure(figsize=(10, 6))
plt.scatter(counties, avg_success_rates, color='blue', label="States")

x_vals = np.linspace(min(counties), max(counties), 500)
plt.plot(x_vals, inverse_polynomial(x_vals, a, b, c), "g--", label="Inverse Polynomial")

formula_text = f"y = {a:.2f} / (x + {b:.2f}) + {c:.2f}"
plt.text(0.05, 0.95, formula_text, transform=plt.gca().transAxes, fontsize=12, verticalalignment='top', color="green")

plt.xlabel("Number of Counties")
plt.ylabel("Average Success Rate (%)")
plt.title("Inverse Polynomial Fit of Average Success Rate vs. Number of Counties")
plt.legend()
plt.show()
```

```python
#inverse polynomial vs. inverse exponential 
def inverse_exponential(x, a, b, c):
    return a * np.exp(-b * x) + c
def inverse_polynomial(x, a, b, c):
    return a / (x + b) + c

#fit the models 
params_exp, _ = curve_fit(inverse_exponential, counties, avg_success_rates, maxfev=10000)
a_exp, b_exp, c_exp = params_exp

params_poly, _ = curve_fit(inverse_polynomial, counties, avg_success_rates, maxfev=10000)
a_poly, b_poly, c_poly = params_poly

pred_exp = inverse_exponential(np.array(counties), a_exp, b_exp, c_exp)
pred_poly = inverse_polynomial(np.array(counties), a_poly, b_poly, c_poly)

mse_exp = mean_squared_error(avg_success_rates, pred_exp)
mse_poly = mean_squared_error(avg_success_rates, pred_poly)
r2_exp = r2_score(avg_success_rates, pred_exp)
r2_poly = r2_score(avg_success_rates, pred_poly)


print(f"Inverse Exponential Model: MSE = {mse_exp:.4f}, R^2 = {r2_exp:.4f}")
print(f"Inverse Polynomial Model: MSE = {mse_poly:.4f}, R^2 = {r2_poly:.4f}")

plt.figure(figsize=(10, 6))
plt.scatter(counties, avg_success_rates, color='blue', label="Data Points")

x_vals = np.linspace(min(counties), max(counties), 500)
plt.plot(x_vals, inverse_exponential(x_vals, *params_exp), "r--", label="Inverse Exponential Fit")
plt.plot(x_vals, inverse_polynomial(x_vals, *params_poly), "g--", label="Inverse Polynomial Fit")

plt.xlabel("Number of Counties")
plt.ylabel("Average Success Rate (%)")
plt.title("Inverse Exponential vs. Inverse Polynomial Fits")
plt.legend()
plt.show()
```
